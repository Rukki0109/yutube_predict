```python
from googleapiclient.discovery import build
import pandas as pd


channel_id = 'UCusnpkgavQhPV_8e_zEh_jw'

# üîß APIÊé•Á∂ö
youtube = build('youtube', 'v3', developerKey=api_key)

# üì¶ ÂãïÁîªID„ÇíÂèñÂæó
def get_video_ids(channel_id):
    video_ids = []
    next_page_token = None

    while True:
        res = youtube.search().list(
            part='id',
            channelId=channel_id,
            maxResults=50,
            pageToken=next_page_token,
            type='video',
            order='date'
        ).execute()

        for item in res['items']:
            video_ids.append(item['id']['videoId'])

        next_page_token = res.get('nextPageToken')
        if not next_page_token:
            break

    return video_ids

# üìä ÂãïÁîªÊÉÖÂ†±„ÇíÂèñÂæó„Åó„Å¶DataFrameÂåñ
def get_video_details(video_ids):
    videos = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(
            part='snippet,statistics',
            id=','.join(video_ids[i:i+50])
        ).execute()

        for item in res['items']:
            videos.append({
                'videoId': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'publishedAt': item['snippet']['publishedAt'],
                'thumbnail': item['snippet']['thumbnails']['high']['url'],
                'viewCount': int(item['statistics'].get('viewCount', 0))
            })
    return pd.DataFrame(videos)

# üì• ÂÆüË°å„Åó„Å¶‰øùÂ≠ò
video_ids = get_video_ids(channel_id)
df = get_video_details(video_ids)
df.to_excel('/data/sutegoro_videos0605.xlsx', index=False)
print("‚úÖ „Éá„Éº„ÇøÂèñÂæóÂÆå‰∫ÜÔºÅ'sutegoro_videos.xlsx' „Å´‰øùÂ≠ò„Åó„Åæ„Åó„Åü„ÄÇ")
```

```python
import pandas as pd
import fugashi
from sklearn.preprocessing import Binarizer
from scipy.stats import pointbiserialr

# ÂΩ¢ÊÖãÁ¥†Ëß£ÊûêÂô®ÔºàfugashiÔºâ„Çí‰Ωø„Å£„Å¶„ÄÅÂêçË©ûÊäΩÂá∫
tagger = fugashi.Tagger()
titles = df["title"].dropna()

# ÂçòË™û„Åî„Å®„ÅÆÂá∫ÁèæÊÉÖÂ†±„Çí„Éê„Ç§„Éä„É™„Åß‰øùÊåÅ
word_occurrences = {}
for i, title in enumerate(titles):
    words = [word.surface for word in tagger(title) if word.feature.pos1 in ['ÂêçË©û']]
    for word in set(words):  # ÈáçË§áÈô§Â§ñ„Åó„Å¶1ÂãïÁîª1„Ç´„Ç¶„É≥„Éà
        if word not in word_occurrences:
            word_occurrences[word] = [0] * len(titles)
        word_occurrences[word][i] = 1

# „Éá„Éº„Çø„Éï„É¨„Éº„É†Âåñ
binary_df = pd.DataFrame(word_occurrences)

# „Ç§„É≥„Éá„ÉÉ„ÇØ„ÇπÊèÉ„Åà„Çã
binary_df.index = titles.index

# ÂêÑÂçòË™û„Å®ÂÜçÁîüÂõûÊï∞„ÅÆÁõ∏Èñ¢„ÇíË®àÁÆó
correlations = []
for word in binary_df.columns:
    try:
        corr, p = pointbiserialr(binary_df[word], df.loc[binary_df.index, "viewCount"])
        if not pd.isna(corr):
            correlations.append((word, corr, p))
    except Exception:
        continue

# Áõ∏Èñ¢„ÅÆÈ´ò„ÅÑÈ†Ü„Å´„ÇΩ„Éº„Éà
correlations_sorted = sorted(correlations, key=lambda x: abs(x[1]), reverse=True)

# ‰∏ä‰Ωç20ÂçòË™û„ÇíË°®Á§∫
top_corr_df = pd.DataFrame(correlations_sorted[:20], columns=["ÂçòË™û", "Áõ∏Èñ¢‰øÇÊï∞", "pÂÄ§"])
top_corr_df
```

```python
import numpy as np

def predict_view_count(title: str, model, top_words):
    # ÁâπÂæ¥Èáè„Éô„ÇØ„Éà„É´„ÇíÊßãÁØâ
    features = []
    for word in top_words:
        features.append(int(word in title))
    
    # XGBoost„ÅØ2Ê¨°ÂÖÉÈÖçÂàó„ÇíË¶ÅÊ±Ç
    X_input = np.array([features])
    
    # log„Çπ„Ç±„Éº„É´„Åß‰∫àÊ∏¨ ‚Üí ÂÖÉ„ÅÆÂÜçÁîüÊï∞„Å´Êàª„Åô
    log_pred = model.predict(X_input)[0]
    view_pred = int(np.expm1(log_pred))
    
    return view_pred
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# üî§ 1. „Çø„Ç§„Éà„É´„ÅÆTF-IDFÁâπÂæ¥ÈáèÔºà100Ê¨°ÂÖÉÔºâ
vectorizer = TfidfVectorizer(max_features=100)
X_tfidf = vectorizer.fit_transform(df["title"].fillna("")).toarray()
tfidf_df = pd.DataFrame(X_tfidf, columns=vectorizer.get_feature_names_out())
tfidf_df.index = df.index

# üî† 2. ÂçòË™û„Éê„Ç§„Éä„É™ÁâπÂæ¥ÈáèÔºà20Ë™ûÔºâ
top_words = [word for word, _, _ in correlations_sorted[:20]]
for word in top_words:
    df[f"has_{word}"] = df["title"].str.contains(word, na=False).astype(int)
binary_features = [f"has_{word}" for word in top_words]

# üî¢ 3. ÊßãÈÄ†„ÉªÊäïÁ®øÊÉÖÂ†±„ÉªË™¨ÊòéÊñá
df["title_len"] = df["title"].str.len()
df["title_word_count"] = df["title"].str.split().str.len()
df["title_has_!"] = df["title"].str.contains("!").astype(int)
df["title_has_Ôºü"] = df["title"].str.contains("Ôºü|\?").astype(int)
df["publishedAt"] = pd.to_datetime(df["publishedAt"])
df["hour"] = df["publishedAt"].dt.hour
df["day_of_week"] = df["publishedAt"].dt.dayofweek
df["desc_exists"] = df["description"].notnull().astype(int)
df["desc_len"] = df["description"].fillna("").str.len()

meta_features = [
    "title_len", "title_word_count", "title_has_!", "title_has_Ôºü",
    "hour", "day_of_week", "desc_exists", "desc_len"
]

# ‚úÖ „Åô„Åπ„Å¶„ÅÆÁâπÂæ¥Èáè„ÇíÁµêÂêà
X = pd.concat([df[binary_features + meta_features], tfidf_df], axis=1)
y = df["log_view"]
```
